```python
import pandas as pd
import numpy as np
import glob
```

```python
bka_files = glob.glob('../data/bka_data/*.xlsx')
bka_files.sort()
for f in bka_files:
    _df = pd.read_excel(f)
    # get period end date
    end_date = pd.to_datetime(_df.iloc[1]['Table 01'].split('Period: ')[1].split(' until ')[1], dayfirst = True).date()
    # create temporary to extract/create column names
    tmp = _df.iloc[2:6].T.copy()
    tmp[2] = tmp[2].ffill()
    tmp.fillna('', inplace = True)
    tmp.replace({'from column 16': 'suspected ',
                 'from column 3\nattempts': 'attempted ',
                 'weiblich': '', 'distribution on crime scenes':
                 'pop dist '}, inplace = True)
    # build column name list
    col_list = tmp[2].map(str) + ' ' + tmp[3].map(str) + ' ' + tmp[4].map(str) + ' ' + tmp[5].map(str)
    col_list = col_list.str.strip().str.split().map(lambda x: ' '.join(x))
    col_list = col_list.str.replace('firarm', 'firearm')
    # get first row with ------
    rows_to_skip = _df[_df[_df.columns[0]].str.contains('-').fillna(False)].index[0]
    # pull and skip rows
    _df = pd.read_excel(f, skiprows = rows_to_skip)
    # add new column name
    _df.columns = col_list
    _df['ref date'] = end_date
    display(_df.head(10))
```
